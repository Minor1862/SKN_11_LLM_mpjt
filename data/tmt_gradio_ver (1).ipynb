```python
import gradio as gr
import ollama
import random
import time
import re

# ëŒ€í™” í‚¤ì›Œë“œ ìƒì„±
def get_random_topic():
    prompt = """
    ì‚¬ìš©ìê°€ ë°•ì°¬í˜¸ì™€ ëŒ€í™”ë¥¼ ì‹œì‘í•˜ë ¤ê³  í•©ë‹ˆë‹¤.
    ì˜¤ëŠ˜ ëŒ€í™” ì£¼ì œë¡œ ì‚¼ì„ë§Œí•œ, ì¼ìƒì ì´ê³  ì¬ë¯¸ìˆëŠ” ë‹¨ì–´ í•˜ë‚˜ë§Œ ê³¨ë¼ì£¼ì„¸ìš”.
    ëŒ€ë‹µì€ ë‹¨ì–´ í•˜ë‚˜ë§Œ í•´ì£¼ì„¸ìš”. ì„¤ëª…ì€ í•„ìš” ì—†ìŠµë‹ˆë‹¤.
    ì˜ˆ: ëƒ‰ë©´, ë¹¨ë˜, ì‹ ë°œ, ê³µí•­, ì‚¬ìš°ë‚˜ ë“±ë“±
    """
    response = ollama.chat(
        model="EEVE-Korean-10.8b",
        messages=[
            {"role": "system", "content": "ë‹¹ì‹ ì€ ì°½ì˜ì ì´ê³  ì¬ì¹˜ìˆëŠ” ë‹¨ì–´ë¥¼ ì¶”ì²œí•´ì£¼ëŠ” ì¡°ìˆ˜ì…ë‹ˆë‹¤. ë°˜ë“œì‹œ ë‹¨ì–´ í•˜ë‚˜ë§Œ ëŒ€ë‹µí•˜ì„¸ìš”."},
            {"role": "user", "content": prompt}
        ]
    )
    keyword = response['message']['content'].strip()
    keyword = re.sub(r'[^\uAC00-\uD7A3a-zA-Z0-9]', '', keyword)
    return keyword

# íˆ¬ë¨¸ì¹˜í† ì»¤ ìŠ¤íƒ€ì¼ ì‘ë‹µ ìƒì„±
def generate_long_response(user_input):
    prompt = f"""
    ì‚¬ìš©ìê°€ \"{user_input}\"ì´ë¼ëŠ” í‚¤ì›Œë“œë¡œ ë§ ê±¸ì—ˆì„ë•Œ, ì•ˆë¶€ì¸ì‚¬ë‚˜ ìê¸°ì†Œê°œë¥¼ í¬í•¨í•œ ì„œë¡ ì€ ìƒëµí•˜ê³  \"ì œê°€ LAì— ìˆì„ ë•Œ ë§ì…ë‹ˆë‹¤â€¦\" í˜¹ì€ \"1994ë…„ ì œê°€ LA ë‹¤ì €ìŠ¤ì—ì„œ...\" ìœ¼ë¡œ ëŒ€í™”ë¥¼ ì‹œì‘í•˜ì„¸ìš”. 
    ëŒ€í™” ë‚´ìš©ì€ ì¤‘ìš”í•˜ì§€ ì•Šì§€ë§Œ ì§„ì§€í•˜ê²Œ ë°•ì°¬í˜¸ íŠ¹ìœ ì˜ íˆ¬ë¨¸ì¹˜í† ì»¤ ë§íˆ¬ë¡œ ì´ì–´ê°€ì„¸ìš”. 
    ì¸ì‚¬ë§ì´ë‚˜ ìì‹ ì„ ì†Œê°œí•˜ëŠ” ì´ì•¼ê¸°ëŠ” ì ˆëŒ€ë¡œ ë“¤ì–´ê°€ì„œëŠ” ì•ˆë©ë‹ˆë‹¤.
    """
    response = ollama.chat(
        model="EEVE-Korean-10.8b",
        messages=[
            {"role": "system", "content": "ë‹¹ì‹ ì€ íˆ¬ë¨¸ì¹˜í† ì»¤ë¡œ ë„ë¦¬ ì•Œë ¤ì§„ ì•¼êµ¬ì„ ìˆ˜ ë°•ì°¬í˜¸ì…ë‹ˆë‹¤. ì¸ì‚¬ë§ì€ ìƒëµí•˜ê³  LAì‹œì ˆ ê²ªì—ˆë˜ ì´ì•¼ê¸°ë¥¼, 10ë¬¸ë‹¨ ì´ìƒìœ¼ë¡œ ê¸¸ê³  ì¥í™©í•˜ê²Œ ë§í•´ì•¼ í•©ë‹ˆë‹¤."},
            {"role": "user", "content": prompt}
        ]
    )
    return response['message']['content']

# í€´ì¦ˆ ìƒì„±
def generate_question_from_text(long_text):
    prompt = f"""
    ì•„ë˜ì˜ ë°•ì°¬í˜¸ ìŠ¤íƒ€ì¼ ê¸´ ëŒ€í™” ë‚´ìš©ì„ ë°”íƒ•ìœ¼ë¡œ, ì¬ë¯¸ìˆëŠ” í€´ì¦ˆë¥¼ í•˜ë‚˜ ë§Œë“¤ì–´ ì£¼ì„¸ìš”.
    â— ì¶œë ¥ í˜•ì‹ì€ ë°˜ë“œì‹œ ì•„ë˜ ì˜ˆì‹œë¥¼ ë”°ë¼ì•¼ í•©ë‹ˆë‹¤:
    ì§ˆë¬¸: (ì—¬ê¸°ì— ì§ˆë¬¸ ì‘ì„±) / ì •ë‹µ: (ì—¬ê¸°ì— ì •ë‹µ ì‘ì„±)
    ëŒ€í™” ë‚´ìš©:
    '''{long_text}'''
    """
    response = ollama.chat(
        model="EEVE-Korean-10.8b",
        messages=[
            {"role": "system", "content": "ë‹¹ì‹ ì€ í˜•ì‹ì„ ì—„ê²©íˆ ì§€í‚¤ëŠ” í€´ì¦ˆ ìƒì„± ì „ë¬¸ê°€ì…ë‹ˆë‹¤. ë°˜ë“œì‹œ 'ì§ˆë¬¸: ~ / ì •ë‹µ: ~' í˜•ì‹ì„ ë”°ë¼ì•¼ í•©ë‹ˆë‹¤."},
            {"role": "user", "content": prompt}
        ]
    )
    q_text = response['message']['content']
    match = re.search(r"ì§ˆë¬¸:\s*(.*?)\s*/\s*ì •ë‹µ:\s*(.*)", q_text)
    if match:
        return match.group(1).strip(), match.group(2).strip()
    else:
        return None, None

# ê²Œì„ ë¡œì§ í•¨ìˆ˜ë“¤
def game_step(user_text, state):
    long_text = generate_long_response(user_text)
    question, answer = generate_question_from_text(long_text)
    if not question or not answer:
        question, answer = "ë°•ì°¬í˜¸ê°€ ìì£¼ ì–¸ê¸‰í•˜ëŠ” ë„ì‹œ ì´ë¦„ì€?", "LA"

    game_log = f"ğŸ§”ğŸ»â€â™‚ï¸ íˆ¬ë¨¸ì¹˜í† ì»¤: {long_text}\n\nâ“ ì§ˆë¬¸: {question}\nì •ë‹µì„ ì…ë ¥í•˜ì„¸ìš”!"
    return game_log, {"question": question, "answer": answer, "long_text": long_text}

def check_answer(user_input, state):
    correct_answer = state.get("answer")
    if user_input.strip().lower() == correct_answer.strip().lower():
        return print("ì •ë‹µì…ë‹ˆë‹¤!ğŸ‰ğŸ‰ íˆ¬ë¨¸ì¹˜í† ì»¤ê°€ ë‹¹ì‹ ì„ ê·¸ë§Œ ë†“ì•„ì¤ë‹ˆë‹¤!(ë”í™©ì± ğŸƒğŸ»â€â™‚ï¸ğŸƒğŸ»â€â™‚ï¸)\n")
    else:
        # ì‹¤íŒ¨ ì‹œ ë¬¸ì¥ ì¤‘ í•˜ë‚˜ë¡œ ë²Œì¹™ íƒ€ì
        long_text = state.get("long_text")
        sentences = re.split(r'(?<=[.!?])\s+', long_text.strip())
        sentences = [s for s in sentences if len(s.strip()) > 10]
        punishment_text = random.choice(sentences).strip() if sentences else long_text.strip()
        punishment_msg = f"âŒ ì˜¤ë‹µì…ë‹ˆë‹¤. ì •ë‹µì€ '{correct_answer}'ì…ë‹ˆë‹¤.\n\nğŸ’¢íˆ¬ë¨¸ì¹˜í† ì»¤ê°€ í™”ë‚¬ìŠµë‹ˆë‹¤!\nì•„ë˜ ë¬¸ì¥ì„ 30ì´ˆ ì•ˆì— ì…ë ¥í•˜ì„¸ìš”!\n\nğŸ’¬ '{punishment_text}'"
        return punishment_msg, {}

# Gradio UI
def play_game(user_text, state):
    if state.get("answer") is None:
        return game_step(user_text, state)
    else:
        return check_answer(user_text, state)

with gr.Blocks() as demo:
    state = gr.State(value={})

    gr.Markdown("# íˆ¬ë¨¸ì¹˜í† ì»¤ì™€ ëŒ€í™”í•˜ê¸° ğŸ’¬ ")
    output_text = gr.Textbox(label="ğŸ§¾ ê²Œì„ ì§„í–‰ ë‚´ìš©", lines=15)
    input_text = gr.Textbox(label="ğŸ—£ï¸ ëŒ€í™”ë¥¼ ì‹œì‘í•˜ê±°ë‚˜ ì •ë‹µì„ ì…ë ¥í•˜ì„¸ìš”")
    submit_button = gr.Button("ğŸ® ê²Œì„ ì‹œì‘ | ğŸ’¬ í…ìŠ¤íŠ¸ ì „ì†¡")

    submit_button.click(play_game, inputs=[input_text, state], outputs=[output_text, state])

if __name__ == '__main__':
    demo.launch(share=True)
```
